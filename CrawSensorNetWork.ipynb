# Craw Sensor Network 

```python
from selenium import webdriver
import requests_html
import pandas as pd
import re, time
```

首先我們觀察這個動態網站, 發現它把東西都藏起來, 直接用requests取資料會拿不到.  
因此偷偷使用selenium來拿網頁原始碼

```python
# 開啟瀏覽器並連到指定的網頁
driver = webdriver.Chrome()
driver.get('https://www.wunderground.com/history/daily/cn/%E4%B8%8A%E6%B5%B7/ZSSS/date/2018-11-3?cm_ven=localwx_history')

# 稍微等一下瀏覽器把網頁載入
time.sleep(5)
# 把東西塞到自己常用的套件中, 如果習慣用beautiful soup的人也可以把資料塞bs4
response = requests_html.HTML(html=driver.page_source)
driver.close()
```

```python
# 選出我們想要解析的表格
table = response.find('#history-observation-table')[0].raw_html.decode()
```

```python
# 抓出表格的columns
a = requests_html.HTML(html=re.findall('<thead.+?</thead>', table, re.S)[0])
columns = [th.text for th in a.find('th')]
```

```python
# 抓出表格的內容
a = requests_html.HTML(html=re.findall('<tbody.+?</tbody>', table, re.S)[0])
data = [[td.text for td in tr.find('td')] for tr in a.find('tr')]
```

```python
# 將資料做成dataframe格式
pd.DataFrame(columns=columns, data=data)
```
